```python
def impute_mediam(series):
    return series.fillna(series.median())
```

```python
df.Year = df['Year'].transform(impute_mediam)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.plot(kind= 'box' , subplots=True, layout=(3,3),
             sharex=False, sharey=False, figsize=(10,8))
```

```python
dfg = df.groupby('Genre')
x = dfg['NA_Sales'].agg(np.mean)
y = dfg['EU_Sales'].agg(np.sum)
z = dfg['JP_Sales'].agg(np.mean)
t = dfg['Global_Sales'].agg(np.mean)
print(x)
print(y)
print(z)
print(t)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(16,5))
plt.plot(x, color='r')
plt.xticks(rotation=90)
plt.title('Genre wise NA_Sales')
plt.xlabel('Genre-->')
plt.ylabel('NA_Sales-->')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(y,'bs', color='b')
plt.xticks(rotation=90)
plt.title('Genre wise EU_Sales')
plt.xlabel('Genre-->')
plt.ylabel('EU_Sales-->')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(z,'r--', color='g')
plt.xticks(rotation=90)
plt.title('Genre wise JP_Sales')
plt.xlabel('Genre-->')
plt.ylabel('JP_Sales-->')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(t,'*--', color='black')
plt.xticks(rotation=90)
plt.title('Genre wise Global_Sales')
plt.xlabel('Genre-->')
plt.ylabel('Global_Sales-->')
plt.show()
```
